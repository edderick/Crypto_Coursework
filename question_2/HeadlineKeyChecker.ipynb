```python
def rotate(s):
    return s[1:]+s[:1]

def sub(ciphertext, key, rotated_key):
    m = {}
    for c, p in zip(key, rotated_key):
        m[p] = c
    
    plaintext = ""
    for c in ciphertext:
        if c == ' ':
            plaintext += ' '
        else:
            plaintext += m[c]
    
    return "{} -- {}".format(plaintext, rotated_key)
    

def check_key(ciphertext, key):
    
    k = key
    
    for i in range(0, len(key)):
        k = rotate(k)
        
        print sub(ciphertext, key, k)

```
